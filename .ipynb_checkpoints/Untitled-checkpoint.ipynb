```python
# importing all the necessary libraries
from playsound import playsound
import time
import matplotlib.pyplot as plt
import librosa
import numpy
```

```python
# playing the sounds
song_li = ["s5_synthetic.wav","beep_fs_10000.wav","s5.wav", "beep_fs_16000.wav", "s5_synthetic.wav"]
for i in song_li:
    playsound(i)
    time.sleep(1)
```

```python
# defining the signum funcion
def sgn(x):
    y = numpy.zeros_like(x)
    y[numpy.where(x > 0)] = 1.0 
    y[numpy.where(x < 0)] = -1.0 
    return y
```

```python
# feature extraction and plotting
li = ["s5.wav","should.wav"]
for i in range(2): 
    wav, fs = librosa. load(li[i])
    n_wav = librosa.util.normalize(wav)
    win = numpy.hamming(max(1, len(n_wav)//64)) 
    win = win/len(win)
    mag = numpy.convolve(numpy.abs(n_wav),win, mode="same")*200
    eng = numpy.convolve(numpy.square(n_wav), numpy.square(win),mode = "same")*2000 
    x = numpy.roll(n_wav, 1)
    x[0] = 0.0 
    abs_diff = numpy.abs(sgn(n_wav) - sgn(x)) 
    zc = numpy.convolve(abs_diff, win, mode="same")*0.5
    fig, ax = plt.subplots(4,constrained_layout=True)
    fig.suptitle("waveform" + str(i+1))
    ax[0].set_title(li[i])
    ax[0].plot(wav)
    ax[1].set_title("STM")
    ax[1].plot(mag)
    ax[2].set_title("STE")
    ax[2].plot(eng)
    ax[3].set_title("ZCR")
    ax[3].plot(zc)
plt.show()
```
